### 1) Use the covid-19 dataset, consider the location attributes like latitude and longitude.Using the Cured cases of the location draw contours. In the cartograph, show xlabel, ylabel, title  and contour value.  

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from matplotlib.colors import Normalize
from matplotlib.colorbar import ColorbarBase

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
contour = ax.tricontourf(data['Longitude'], data['Latitude'], data['Cured/Discharged/Migrated'], 
                         transform=ccrs.PlateCarree(), cmap='coolwarm')  # colormap as 'coolwarm'
cbar = plt.colorbar(contour, orientation='vertical', fraction=0.046, pad=0.04)
cbar.set_label('Cured Cases')

ax.set_xlabel('Longitude',labelpad=15)
ax.set_ylabel('Latitude',labelpad=15)
ax.set_title('Contour for Cured Cases')
ax.gridlines(draw_labels=True)

plt.show()
```

### 2) For covid19 indian dataset , using Python code with proper visual idioms display the gird chart with 2X3 consists of scatter plot, line plot with point and line, vertical bar plot, horizontal bar plot, box plot and 3D Pie chart.

```python
data1=pd.read_csv('covid.csv')
data1
```

```python
plt.figure(figsize=(30,10))

# Scatter plot
plt.subplot(2,3,1)
plt.scatter(data1['state'],data1['deaths'],marker='.',color='blue',sizes=(600,200,200,500,200,400))
plt.xlabel('States')
plt.ylabel("No.of.Death")
plt.title("No of Deaths over State")

# Line plot
plt.subplot(2,3,2)
plt.plot(data1['state'],data1['deaths'],marker='*',ls=':',mec='orange',mfc='black',ms='20')
plt.xlabel('States')
plt.ylabel("No.of.Death")
plt.title("No of Deaths over State")

# Vertical bar plot
plt.subplot(2,3,3)
plt.bar(data1['state'],data1['deaths'],color='violet')
plt.xlabel('States')
plt.xticks(rotation=75)
plt.ylabel("No.of.Death")
plt.title("No of Deaths over State")

# Horizontal bar plot
plt.subplot(2,3,4)
plt.barh(data1['state'],data1['deaths'],color='violet')
plt.ylabel('States')
plt.xlabel('Deaths')
plt.title("No of Deaths over State")

# Box plot
plt.subplot(2,3,5)
plt.boxplot(data1['deaths'])
plt.title("Box Plot for Deaths")

# 3D Pie chart
plt.subplot(2,3,6)
plt.pie(data1['deaths'].head(5),labels=data1['state'].head(5),autopct='%.2f',shadow='True')
plt.title("Pie Chart ")
```

### 3) For the given train network data between the station and the train type as Super Fast/ Express is given in the table. Design visual idioms using force directed placement method and vertical top down placement method. Apply color and other channels to encode the distance and train type attributes in the idioms. Distinguish longest and shortest paths through appropriate visual encodings.


```python
import networkx as nx
import matplotlib.pyplot as plt

train_data = [
    ("CAPE", "TVC", 80, "SF"),
    ("TVC", "CBE", 425, "SF"),
    ("CBE", "KPD", 400, "SF"),
    ("KPD", "CTO", 35, "EXP"),
    ("KPD", "BZA", 450, "SF"),
    ("BZA", "GNT", 35, "EXP"),
    ("BZA", "TEL", 40, "EXP"),
    ("BZA", "NGP", 700, "SF"),
    ("NGP", "SBC", 1092, "SF"),
    ("SBC", "MAQ", 370, "SF"),
    ("CBE", "TVC", 425, "SF"),
    ("KPD", "CBE", 400, "SF"),
    ("BZA", "KPD", 450, "SF"),
    ("NGP", "BZA", 700, "SF"),
    ("SBC", "NGP", 1092, "SF"),
    ("CTO", "TPTY", 70, "EXP"),
    ("CTO", "MLP", 95, "EXP"),
    ("MAQ", "UD", 50, "EXP")
]

G = nx.DiGraph()

for source, destination, distance, express_type in train_data:
    G.add_edge(source, destination, distance=distance, express_type=express_type)

min_distance = min(nx.get_edge_attributes(G, 'distance').values())
max_distance = max(nx.get_edge_attributes(G, 'distance').values())
edge_colors = ['violet' if G.edges[e]['distance'] == min_distance else
               'orange' if G.edges[e]['distance'] == max_distance else
               'blue' for e in G.edges]

plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_size=700, font_size=10, font_weight='bold',
                 edge_color=edge_colors, width=4)
plt.title('Force-Directed Graph of Train Network')
plt.show()

plt.figure(figsize=(10, 6))
pos = nx.drawing.layout.planar_layout(G, scale=1000)
nx.draw_networkx(G, pos, with_labels=True, node_size=700, font_size=10, font_weight='bold',
                 edge_color=edge_colors, width=4)
plt.title('Vertical Top-Down Tree Diagram of Train Network')
plt.show()
```

# 4) Using Python code present the statewise covid-19 data using choropleth map and pointmap .

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# Load the data
# Replace 'data' with your actual DataFrame containing COVID-19 cases data

# Create a choropleth map for India
fig_choropleth = px.choropleth(
    data,
    geojson='https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson',
    featureidkey='properties.ST_NM',
    locations='Name of State / UT',
    color='New cases',
    hover_name='Name of State / UT',
    color_continuous_scale='cividis',  # Changed color scale to 'cividis'
    title='COVID-19: Statewise New Cases in India',
)

fig_choropleth.update_geos(projection_type="mercator")

# Modify title, axis names, and colors
fig_choropleth.update_layout(
    title_text='Statewise Distribution of New COVID-19 Cases in India using choropleth',
    coloraxis_colorbar_title='New Cases',
    coloraxis_colorbar_ticks='outside',
    coloraxis_colorbar_tickvals=[0, 1000, 2000, 3000],  # Modify colorbar tick values
    coloraxis_colorbar_ticktext=['0', '1K', '2K', '3K'],  # Modify colorbar tick labels
    margin=dict(l=0, r=0, t=50, b=0)  # Adjust layout margins
)

# Create a point map for India
fig_pointmap = go.Figure(go.Scattergeo(
    lon=data['Longitude'],
    lat=data['Latitude'],
    text=data['Name of State / UT'] + '<br>Cases: ' + data['New cases'].astype(str),
    mode='markers',
    marker=dict(
        size=data['New cases'] * 0.1,
        opacity=0.8,
        reversescale=True,
        autocolorscale=False,
        symbol='circle',
        line=dict(
            width=1,
            color='rgba(102, 102, 102)'
        ),
        colorscale='cividis',  # Changed color scale to 'cividis'
        cmin=0,
        color=data['New cases'],
        cmax=data['New cases'].max(),
        colorbar_title='New Cases'
    )
))

# Modify title, axis names, and colors
fig_pointmap.update_layout(
    title='COVID-19 Cases in each State using Point Map',
    geo=dict(
        showcoastlines=True,
        projection_type='natural earth'
    ),
    xaxis_title='Longitude',
    yaxis_title='Latitude',
    coloraxis_colorbar_title='New Cases',
    coloraxis_colorbar_ticks='outside',
    coloraxis_colorbar_tickvals=[0, 1000, 2000, 3000],  # Modify colorbar tick values
    coloraxis_colorbar_ticktext=['0', '1K', '2K', '3K'],  # Modify colorbar tick labels
)

# Show the maps
fig_choropleth.show()
fig_pointmap.show()
```
