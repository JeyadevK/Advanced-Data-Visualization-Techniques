```python
a=covid.state
b=covid.active
c=covid.deaths
plt.figure(figsize=(16,14))
plt.subplot(2, 1, 1)
plt.bar(a,b,color='red')
plt.xlabel("State",color='brown')
plt.xticks(rotation=75)
plt.ylabel("Active Cases",color='brown')
plt.title("No of Active Cases over state")

plt.figure(figsize=(16, 14))
plt.subplot(2, 1, 2)
plt.bar(a,c,color='Blue')
plt.xlabel("State",color='brown')
plt.xticks(rotation=75)
plt.ylabel("No.of.Death",color='brown')
plt.title("No of Deaths over State")
```

5.	Apply horizontal and vertical align on one attribute

```python
x=covid.state
y=covid.dose1
plt.figure(figsize=(16,14))
plt.subplot(2, 1, 1)
plt.bar(x,y,color='violet')
plt.xlabel("State",color='brown')
plt.xticks(rotation=75)
plt.ylabel("Dose 1",color='brown')

plt.figure(figsize=(14, 17))
plt.subplot(2, 1, 2)
plt.barh(x,y,color='Blue')
plt.xlabel("Dose 1",color='brown')
plt.xticks(rotation=75)
plt.ylabel("State",color='brown')
```

6.	Apply ascending and descending ordering one another attribute

```python
ascen=covid.sort_values(by='dose2',ascending=True)
plt.figure(figsize=(16,14))
plt.bar(ascen['state'],ascen['dose2'],color='yellow')
plt.xlabel("State")
plt.xticks(rotation=75)
plt.ylabel("Dose 2")
plt.title("State vs Dose 2 in Ascending order")
```

```python
ascen=covid.sort_values(by='dose2',ascending=False)
plt.figure(figsize=(16,14))
plt.bar(ascen['state'],ascen['dose2'],color='yellow')
plt.xlabel("State")
plt.xticks(rotation=75)
plt.ylabel("Dose 2")
plt.title("State vs Dose 2 in Descending order")
```

Exercise 3

 For the covid19 data set, consider one of the quantitative attribute. Use two hue, luminescent and rainbow colormap to present the data. Include title, labels and colorbar for visual presentation. Write the observation made out of the presented values using colormaps.


```python
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.colors import LinearSegmentedColormap
```

```python
covid=pd.read_csv("covid.csv")
covid
```

```python
#Two hue Colormap
colors = [(0, 0, 1), (1, 0.5, 0)]  # Blue to Orange
cmap_name = 'two_hue'
two_hue_cmap = LinearSegmentedColormap.from_list(cmap_name, colors)

plt.figure(figsize=(8, 6))
plt.scatter(covid['state'].head(10),covid['active'].head(10), c=covid['dose1'].head(10), cmap=two_hue_cmap, alpha=0.8)
plt.colorbar(label='Intensity')
plt.xlabel("State")
plt.xticks(rotation=75)
plt.ylabel("Active Cases")
plt.title('Two-Hue Colormap',color='blue')

plt.show()
```

```python
# Rainbow colormap
plt.figure(figsize=(12, 6))
plt.scatter(covid['state'].head(10),covid['active'].head(10),
            c=covid['dose1'].head(10), cmap='rainbow', alpha=0.8)
plt.colorbar(label='Intensity')
plt.xlabel("State")
plt.xticks(rotation=75)
plt.ylabel("Active Cases")
plt.title('Rainbow Colormap',color='violet')
```

```python
#Luminance colormap
plt.figure(figsize=(12, 6))
plt.scatter(covid['state'].head(10),covid['active'].head(10), c=covid['dose1'].head(10), cmap='plasma', alpha=0.8)
plt.colorbar(label='Intensity')
plt.xlabel("State")
plt.xticks(rotation=75)
plt.ylabel("Active Cases")
plt.title('Luminance Colormap by Plasma',color='Blue')
```
